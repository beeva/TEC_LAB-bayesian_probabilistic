### Loss Error Experiments

The goal of these experiments is to understand the dependency of the original loss function in modelling the variance of the error in the prediction. We try to validate the following hyphotesis:

### Hypothesis

> **H1**: Do we have the same result if we change the loss function from case a to case b?
- a) - Original Exp.1 - 1 input tensor: [y_hat, sigma] and 1 loss error 
- b) - Common Usage - 2 input tensors: [y_hat], [sigma] and 1 loss error (loss_yhat + loss_sigma)

*We will try to validate both cases in order to propose the most common and compatible version of this technique*

> **H2**: What is the effect of each loss error (sigma and y) in updating the weights of the network?*

*We will try to understand what is the effect of updating and propagating sigma*

> **H3**: What happens if we only propagate the y error while keeping updating the sigma on the fly?*

*We will try to understand what is the effect of each loss separetly sigma and y in training*

```python
def weight_reset(m):
    if  isinstance(m, torch.nn.Linear):
        m.reset_parameters()
```

### Imports 

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch.autograd import Variable
import torch.nn.functional as F

%matplotlib inline
```

### Generate some random data

```python
# Generate some linearly related data
n_samples = 10000
x = np.random.random(n_samples)
eps = np.random.normal(scale=1.0, size=n_samples) * 0.4*np.sin(x*4*np.pi)
y = 2*x + 1 + eps
```

```python
# Plot the data
plt.figure(figsize=(14,6))
plt.plot(x, y, '.')
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
```

```python
# Add in column of dummy zeros to represent the variance 
var = np.zeros(10000)
y = np.stack([y, var], axis=1)
```

```python
# Reshape x for PyTorch
x = x.reshape(10000, 1)
```

```python
# Split into training and validation sets
n_validation = 500
x_val, x = x[:n_validation], x[n_validation:]
y_val, y = y[:n_validation], y[n_validation:]
```

```python
# Convert x and y to PyTorch Variables on the GPU
dtype = torch.cuda.FloatTensor
x = Variable(torch.from_numpy(x).type(dtype))
y = Variable(torch.from_numpy(y).type(dtype), requires_grad=False)
```

```python
# Check shape ok for PyTorch - x should be (n x 1), y (n x 2)
x.shape, y.shape
```

### Create a simple network 

```python
# Create a simple two-layer network with one input (x) and two outputs (y, sigma)
n_inputs = 1
n_outputs = 2
n_hidden = 1000
model = torch.nn.Sequential(torch.nn.Linear(n_inputs, n_hidden),
                            torch.nn.ReLU(),
                            torch.nn.Linear(n_hidden, n_outputs)
                           ).cuda()
```

```python
# Adam optimizer
learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

### Custom loss function

This is the original implementation. We jointly compute the loss of the target and variance. We return a single scalar

```python
def single_loss_error(input, target):
    # Estimate target value for variance (sigma^2) with (y_pred - y)**2
    #
    #    actual y        is target[:,0]
    # predicted y        is input[:,0]
    #    actual variance is target[:,1] - estimated here
    # predicted variance is input[:,0]
    
    # Use 'requires_grad == False' to prevent PyTorch from trying to differentiate 'target'
    target[:,1] = Variable((input[:,0].data - target[:,0].data)**2, 
                           requires_grad=False)  # compute variance
    # Return MSE loss 
    return F.mse_loss(input, target)
```

We compute separately the loss of the target and the variance   

```python
def multi_loss_error(input, target):
    # Estimate target value for variance (sigma^2) with (y_pred - y)**2
    #
    #    actual y        is target[:,0]
    # predicted y        is input[:,0]
    #    actual variance is target[:,1] - estimated here
    # predicted variance is input[:,0]
    
    # Use 'requires_grad == False' to prevent PyTorch from trying to differentiate 'target'
    target[:,1] = Variable((input[:,0].data - target[:,0].data)**2, 
                           requires_grad=False)  # compute variance
    # Return MSE loss 
    return F.mse_loss(input[:,1], target[:,1]), F.mse_loss(input[:,0], target[:,0])
```

 We only compute the loss of the target. We keep updating sigma on the fly

```python
def single_loss_y_error(input, target):
    # Estimate target value for variance (sigma^2) with (y_pred - y)**2
    #
    #    actual y        is target[:,0]
    # predicted y        is input[:,0]
    #    actual variance is target[:,1] - estimated here
    # predicted variance is input[:,0]
    
    # Use 'requires_grad == False' to prevent PyTorch from trying to differentiate 'target'
    target[:,1] = Variable((input[:,0].data - target[:,0].data)**2, 
                           requires_grad=False)  # compute variance
    # Return MSE loss 
    return F.mse_loss(input[:,0], target[:,0])
```

## Experiments

### Experiment 1 - Loss function - framework compability

> **H1**: Do we have the same result if we change the loss function from case a to case b?
- a) - Single Loss Error - Original Exp.1 - 1 input tensor: [y_hat, sigma] and 1 loss error 
- b) - Common Usage* - 2 input tensors: [y_hat], [sigma] and 1 loss error (loss_yhat + loss_sigma)

Here, we try to validate that we get the same results in case b) as it is found more frequently in most of implementations. The goal is to ensure compability of Tensorflow and Pytorch frameworks

The original implementation of this technique uses the case a) where we have one input tensor of two dimensions. However, we have found that this is not a common implementation. After some research, we have found that case b) is more appropiate and it ensures tensorflow compability 

### Case a) - Single Loss Error - Original

This is the original implementation of the technique. It implements the case a) described before

*Note: this technique is explained in the notebook initial_validation*

```python
# We reset model weights and sigma variance   
model.apply(weight_reset)

y[:,1] = torch.zeros(y.shape[0]).cuda() # reset sigma
```

#### Training loop 
We train this simple model with batch size equal to the whole dataset.

```python
n_epochs = 10000
for i in range(n_epochs):
    
    # Calculate predicted y from x
    y_pred = model(x)
    
    # Calculate loss
    loss = single_loss_error(y_pred, y)
    if i%500 == 0: print(f'epoch: {i:4}, loss: {loss.data.item():.3}',)

    # Backprop, first zeroing gradients
    optimizer.zero_grad()
    loss.backward()

    # Update parameters
    optimizer.step()
```

#### Results

```python
# Get predicted y and its predicted variance for validation set
x_val_pytorch = Variable(torch.from_numpy(x_val).type(dtype)) 
pred = model(x_val_pytorch)
y_pred = pred[:,0].cpu().data
variance_pred = pred[:,1].cpu().data
```

```python
# Plot predictions and their errors
plt.figure(figsize=(14,6))
plt.errorbar(x_val, y_pred, yerr=variance_pred.sqrt() , fmt='.'); 

# One standard-deviation envelope
xs = np.linspace(0.0, 1.0, 50)
err = 0.4*np.sin(xs*4*np.pi)
plt.fill_between(xs, 2*xs+1-err, 2*xs+1+err, facecolor='orange', alpha=0.5, edgecolor='none');
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.title('Validation set predictions');
plt.legend([r'Original data $\pm \sigma$ envelope','Validation set predictions'], loc='upper left', );
```

###  Case B) - Multi Loss Error  - Same Weights

Here, we compute the loss error separetly and return two scalars, one for each. The goal is to ensure compability of Tensorflow and Pytorch frameworks:

 "....except that the losses have to be computed for both outputs and added together before performing backpropagation and model weight updates.." [1]

For the implemenation of this method has been taken as a reference:

[1] https://towardsdatascience.com/converting-a-deep-learning-model-with-multiple-outputs-from-pytorch-to-tensorflow-a2d27a8e44f4

```python
# We reset model weights and sigma variance to avoid mixing experiments
model.apply(weight_reset)
y[:,1] = torch.zeros(y.shape[0]).cuda() # reset sigma
```

#### Training loop 
We train this simple model with batch size equal to the whole dataset.

```python
n_epochs = 10000
for i in range(n_epochs):
    
    # Calculate predicted y from x
    y_pred = model(x)
    
    # Calculate loss
    loss_var, loss_y = multi_loss_error(y_pred, y)
    if i%500 == 0: print(f'epoch: {i:4}, loss: {loss_y.data.item():.3}',)

    # Backprop, first zeroing gradients
    optimizer.zero_grad()
    loss = torch.add(loss_var, loss_y)
    loss.backward()

    # Update parameters
    optimizer.step()
```

#### Results

```python
# Get predicted y and its predicted variance for validation set
x_val_pytorch = Variable(torch.from_numpy(x_val).type(dtype)) 
pred = model(x_val_pytorch)
y_pred = pred[:,0].cpu().data
variance_pred = pred[:,1].cpu().data
```

```python
# Plot predictions and their errors
plt.figure(figsize=(14,6))
plt.errorbar(x_val, y_pred, yerr=variance_pred.sqrt() , fmt='.'); 

# One standard-deviation envelope
xs = np.linspace(0.0, 1.0, 50)
err = 0.4*np.sin(xs*4*np.pi)
plt.fill_between(xs, 2*xs+1-err, 2*xs+1+err, facecolor='orange', alpha=0.5, edgecolor='none');
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.title('Validation set predictions');
plt.legend([r'Original data $\pm \sigma$ envelope','Validation set predictions'], loc='upper left', );
```

As we can see in this plot, we get the same results as in the case a)

#### Conclusions

We get exactly the same results. We have discovered that The Pytorch loss function implementation does this underneath*.
The case b), loss 'multi_loss_error' is the recommended and common implementation for working with multi output tensors 

*The function MSE of Pytorch uses reduce_mean in both dimensions and it return an scalar for doing backpropagation*

### Experiment 2 - Single Loss Error - Only y error 

> **H2**: What happens if we only propagate the y error while keeping updating the sigma on the fly?

Here, we will try to understand what is the effect of updating and propagating sigma

*NOTE: we reset weights to avoid mixing experiments*

```python
def weight_reset(m):
    if  isinstance(m, torch.nn.Linear):
        m.reset_parameters()

model.apply(weight_reset)
y[:,1] = torch.zeros(y.shape[0]).cuda() # reset sigma
```

```python
n_epochs = 10000
for i in range(n_epochs):
    
    # Calculate predicted y from x
    y_pred = model(x)
    
    # Calculate loss
    loss_y = single_loss_y_error(y_pred, y) # We only compute the loss of the target
    if i%500 == 0: print(f'epoch: {i:4}, loss: {loss_y.data.item():.3}',)

    # Backprop, first zeroing gradients
    optimizer.zero_grad()
    loss = loss_y
    
    loss.backward()

    # Update parameters
    optimizer.step()
```

```python
# Get predicted y and its predicted variance for validation set
x_val_pytorch = Variable(torch.from_numpy(x_val).type(dtype)) 
pred = model(x_val_pytorch)
y_pred = pred[:,0].cpu().data
variance_pred = pred[:,1].cpu().data
```

```python
# Plot predictions and their errors
plt.figure(figsize=(14,6))
plt.errorbar(x_val, y_pred, yerr=variance_pred.sqrt(), fmt='.'); 

# One standard-deviation envelope
xs = np.linspace(0.0, 1.0, 50)
err = 0.4*np.sin(xs*4*np.pi)
plt.fill_between(xs, 2*xs+1-err, 2*xs+1+err, facecolor='orange', alpha=0.5, edgecolor='none');
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.title('Validation set predictions');
plt.legend([r'Original data $\pm \sigma$ envelope','Validation set predictions'], loc='upper left', );
```

Here, we can see that this model is not able to predict the real sinusoidal variance, we get almost constant variance of the target $y$ variable along the $x$ axis

The variance that we learn increase slightly, it might be caused due to the effect of loss of $y$ and its increasing trend along the $x$ axis

### Conclusions

We get an homoscedastic variance 

### Experiment 3 - Multi Loss Error - Different Weights - w_sigma, w_y

> * **H3**: What is the effect of each loss error (sigma and y) in updating the weights of the network?

*In this experiment we use the multi_loss_error function defined before to compute separately the loss of the target and the variance. We will play applying different factors to each loss: w_sigma and w_y. We will reduce the value of w_sigma to understand its effect in the backpropagation process*

```python
# we reset weights to avoid mixing experiments
model.apply(weight_reset)
y[:,1] = torch.zeros(y.shape[0]).cuda() # reset sigma
```

**Weight factors** 

Here, we setup the weights factors of sigma and y

We keep the weight of y and we reduce sigma multiplying by a factor of 0.005

```python
w_sigma = 0.005
w_y = 1
```

We train the network using the function defined before 'multi_loss_error'. This function computes separately the loss of the target and the variance   

```python
n_epochs = 10000
for i in range(n_epochs):
    
    # Calculate predicted y from x
    y_pred = model(x)
    
    # Calculate loss
    loss_var, loss_y = multi_loss_error(y_pred, y)
    if i%500 == 0: print(f'epoch: {i:4}, loss: {loss_y.data.item():.3}',)

    # Backprop, first zeroing gradients
    optimizer.zero_grad()
    loss = torch.add(loss_var * w_sigma, loss_y * w_y)
    loss = loss 
    
    loss.backward()

    # Update parameters
    optimizer.step()
```

```python
# Get predicted y and its predicted variance for validation set
x_val_pytorch = Variable(torch.from_numpy(x_val).type(dtype)) 
pred = model(x_val_pytorch)
y_pred = pred[:,0].cpu().data
variance_pred = pred[:,1].cpu().data
```

```python
# Plot predictions and their errors
plt.figure(figsize=(14,6))
plt.errorbar(x_val, y_pred, yerr=variance_pred.sqrt() , fmt='.'); 

# One standard-deviation envelope
xs = np.linspace(0.0, 1.0, 50)
err = 0.4*np.sin(xs*4*np.pi)
plt.fill_between(xs, 2*xs+1-err, 2*xs+1+err, facecolor='orange', alpha=0.5, edgecolor='none');
plt.xlabel(r'$x$');
plt.ylabel(r'$y$');
plt.title('Validation set predictions');
plt.legend([r'Original data $\pm \sigma$ envelope','Validation set predictions'], loc='upper left', );
```

#### Conclusions

We see that the y loss dominates the backpropagation in training. We only need to propagate a small portion of the variance loss to get almost the same effect. This is due to the dependency of the loss variance and the loss of the target
